# Dashboard Comparativo de Búsquedas

Añade en la lista `busquedas` los nombres exactos de las carpetas (o parte del nombre) que se encuentran en `output/` que deseas comparar.
Al ejecutar la celda, se generará un archivo `dashboard_comparativo.html` que podrás abrir en tu navegador para ver los gráficos lado a lado.

```python
# PARAMETROS PARA ELEGIR GRAFICOS
import os
import webbrowser
import base64

# 1. Configuración del Dashboard (Fácilmente escalable)
filtro_busqueda = 'buy_new'  # Cambia a 'buy_new' o deja vacío '' para comparar todo
generar_histogramas = True
generar_scatterplots = False  # Cambia a True si quieres generar el de scatterplots también
```

```python

# 2. Encontrar carpetas coincidentes en output/
output_dir = '../output'
carpetas_encontradas = []

if os.path.exists(output_dir):
    todas_las_carpetas = [d for d in os.listdir(output_dir) if os.path.isdir(os.path.join(output_dir, d))]
    for c in todas_las_carpetas:
        if filtro_busqueda.lower() in c.lower():
            carpetas_encontradas.append(c)

# 3. Función para ordenar (Ciudad alfabético -> 'dentro' primero -> 'afueras'/'fuera' segundo)
def custom_sort(folder_name):
    name_lower = folder_name.lower()
    # Identificar si es dentro o fuera (0 primero, 1 después)
    is_dentro = 0 if 'dentro' in name_lower else 1
    
    # Extraer la ciudad (asumiendo formato ..._Ciudad_zona)
    parts = name_lower.split('_')
    city = parts[3] if len(parts) >= 4 else name_lower
    
    return (city, is_dentro)

# Aplicar el orden personalizado
carpetas_encontradas.sort(key=custom_sort)

print(f"Carpetas encontradas (Ordenadas): {carpetas_encontradas}\n")

def get_base64_image(image_path):
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode('utf-8')
    return f"data:image/png;base64,{encoded_string}"

# Estilos CSS comunes para ambas plantillas
css_styles = """
    <style>
        body { font-family: Arial, sans-serif; background-color: #f4f4f9; color: #333; padding: 20px; }
        h1 { text-align: center; color: #2c3e50; }
        .container { display: flex; flex-direction: column; gap: 40px; margin-top: 30px; }
        .row { display: flex; flex-wrap: wrap; gap: 20px; justify-content: center; }
        .card { background: white; padding: 15px; border-radius: 8px; box-shadow: 0 4px 6px rgba(0,0,0,0.1); text-align: center; max-width: 48%; }
        .card img { max-width: 100%; height: auto; border-radius: 4px; }
        .card h3 { margin-top: 0; font-size: 16px; color: #555; }
    </style>
"""

# 4. Generar HTML Histogramas
if generar_histogramas:
    html_hist = f"<!DOCTYPE html><html lang='es'><head><meta charset='UTF-8'><title>Comparativa Histogramas</title>{css_styles}</head><body>"
    html_hist += f"<h1>Comparativo de Histogramas ({filtro_busqueda})</h1><div class='container'><div class='row'>"
    
    for c in carpetas_encontradas:
        folder_path = os.path.join(output_dir, c)
        img_path = None
        for f in os.listdir(folder_path):
            if f.endswith('.png') and ('sqm_dist' in f.lower() or 'room_dist' in f.lower()):
                img_path = os.path.join(folder_path, f)
                break
        if img_path:
            b64_src = get_base64_image(img_path)
            html_hist += f"<div class='card'><h3>{c}</h3><img src='{b64_src}' alt='Histograma {c}'></div>"
            
    html_hist += "</div></div></body></html>"
    hist_filename = filtro_busqueda+"_histogramas.html"
    with open(hist_filename, "w", encoding="utf-8") as f: f.write(html_hist)
    print(f"Generado: {hist_filename} (Imágenes Base64 incrustadas)")
    webbrowser.open('file://' + os.path.realpath(hist_filename))

# 5. Generar HTML Scatterplots
if generar_scatterplots:
    html_scat = f"<!DOCTYPE html><html lang='es'><head><meta charset='UTF-8'><title>Comparativa Scatterplots</title>{css_styles}</head><body>"
    html_scat += f"<h1>Comparativo Scatterplots ({filtro_busqueda})</h1><div class='container'><div class='row'>"
    
    for c in carpetas_encontradas:
        folder_path = os.path.join(output_dir, c)
        img_path = os.path.join(folder_path, 'precio_vs_tamano.png')
        if os.path.exists(img_path):
            b64_src = get_base64_image(img_path)
            html_scat += f"<div class='card'><h3>{c}</h3><img src='{b64_src}' alt='Scatterplot {c}'></div>"
            
    html_scat += "</div></div></body></html>"
    scat_filename = filtro_busqueda+"_scatterplots.html"
    with open(scat_filename, "w", encoding="utf-8") as f: f.write(html_scat)
    print(f"Generado: {scat_filename} (Imágenes Base64 incrustadas)")
    webbrowser.open('file://' + os.path.realpath(scat_filename))
```
